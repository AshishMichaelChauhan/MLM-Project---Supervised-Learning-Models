```python
def summarize_numerical(my_sample, variables):
    summary = {}

    for col in variables:
        if col in ac07_sample[ac07_non_cat_vars]:  # Check if the column exists in the DataFrame
            summary[col] = {
                'Mode': my_sample[col].mode()[0],
                'Range': my_sample[col].max() - my_sample[col].min(),
                'Skewness': my_sample[col].skew() ,
                'Kurtosis': my_sample[col].kurtosis()
            }
    return summary
```

```python
# Confidence Interval for the Mean
def confidence_interval(data, confidence=0.95):
    mean = data.mean()
    std = data.std()
    n = len(data)
    z = stats.norm.ppf(1 - (1 - confidence) / 2)  # z-score for 95% CI
    margin_of_error = z * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return (lower_bound, upper_bound)

ci_results = {}
for col in ac07_non_cat_vars:
    ci_results[col] = confidence_interval(ac07_sample[col])


print("\nConfidence Intervals (95%):")
for col, ci in ci_results.items():
    print(f"{col}: {ci[0]:.2f} to {ci[1]:.2f}")
```

- This narrow interval suggests high precision in estimating the true population mean of Quantity, indicating a centralized distribution around 0.5.

```python
# Test of Homogeneity {Chi-sq} on categorical data

ac07_contingency_table = pd.crosstab(ac07_sample['Import_Export'], ac07_sample['Category'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(ac07_contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpret the results
ac07_alpha = 0.05  # Significance level

if p < ac07_alpha:
    print("Reject the null hypothesis: There is a significant association between the two categorical variables.")
else:
    print("There is no significant association between the two categorical variables.")
```

```python
# Test of Normality {Shapiro-Wilk | Kolmogorov-Smirnov | Anderson-Darling | Jarque-Bera

def normality_tests(data):

    ac07_results = {}

    # Shapiro-Wilk test
    shapiro_statistic, shapiro_pvalue = shapiro(data)
    ac07_results["Shapiro-Wilk"] = {"statistic": shapiro_statistic, "p-value": shapiro_pvalue}

    # Kolmogorov-Smirnov test (against a normal distribution)
    ks_statistic, ks_pvalue = kstest(data, 'norm')
    ac07_results["Kolmogorov-Smirnov"] = {"statistic": ks_statistic, "p-value": ks_pvalue}

    # Anderson-Darling test
    anderson_result = anderson(data)
    ac07_results["Anderson-Darling"] = {
        "statistic": anderson_result.statistic,
        "critical_values": anderson_result.critical_values,
        "significance_level": anderson_result.significance_level
    }

    # Jarque-Bera test
    jb_statistic, jb_pvalue = jarque_bera(data)
    ac07_results["Jarque-Bera"] = {"statistic": jb_statistic, "p-value": jb_pvalue}

    return ac07_results
```

```python
from scipy import stats

# Test of Correlation (T-test)
def t_test_correlation(data1, data2):
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    return {"t-statistic": t_statistic, "p-value": p_value}

for col in ac07_non_cat_vars:
    print(f"Normality tests for {col}:")
    normality_test_results = normality_tests(ac07_sample[col])
    for test_name, result in normality_test_results.items():
        print(f"  {test_name}: {result}")

    # Example of T-test
    if len(ac07_non_cat_vars) > 1:
        for i in range(len(ac07_non_cat_vars)):
            for j in range(i+1, len(ac07_non_cat_vars)):
                col1, col2 = ac07_non_cat_vars[i], ac07_non_cat_vars[j]
                t_test_result = t_test_correlation(ac07_sample[col1], ac07_sample[col2])
                print(f"T-test Correlation between {col1} and {col2}: {t_test_result}")
    print()
    print('*'*20)
    print()
```

####**Time Taken**

```python
import time
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def time_model_run(model, X_train, y_train):
    start_time = time.time()
    model.fit(X_train, y_train)
    end_time = time.time()
    return end_time - start_time

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55022)
log_reg_time = time_model_run(log_reg_model, ac07_X_train, ac07_y_train)
print(f"Logistic Regression runtime: {log_reg_time:.4f} seconds")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55022)
svm_time = time_model_run(svm_model, ac07_X_train, ac07_y_train)
print(f"SVM runtime: {svm_time:.4f} seconds")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55022)
sgd_time = time_model_run(sgd_model, ac07_X_train, ac07_y_train)
print(f"SGD runtime: {sgd_time:.4f} seconds")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55022)
dt_time = time_model_run(dt_model, ac07_X_train, ac07_y_train)
print(f"Decision Tree runtime: {dt_time:.4f} seconds")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_time = time_model_run(knn_model, ac07_X_train, ac07_y_train)
print(f"KNN runtime: {knn_time:.4f} seconds")

# Naive Bayes
nb_model = GaussianNB()
nb_time = time_model_run(nb_model, ac07_X_train, ac07_y_train)
print(f"Naive Bayes runtime: {nb_time:.4f} seconds")

# Random Forest
rf_model = RandomForestClassifier(random_state=55022)
rf_time = time_model_run(rf_model, ac07_X_train, ac07_y_train)
print(f"Random Forest runtime: {rf_time:.4f} seconds")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55022)
xgb_time = time_model_run(xgb_model, ac07_X_train, ac07_y_train)
print(f"XGBoost runtime: {xgb_time:.4f} seconds")
```

### **Observations:**

- **Fastest:** KNN and Naive Bayes are the fastest, taking approximately 0.02–0.03 seconds.
- **Slowest:** SVM is the slowest, taking over 2.1 seconds.
- **Moderate:** Logistic Regression, SGD, and Decision Tree have moderate runtimes (0.05–0.08 seconds).
- **Relatively Slow:** Random Forest and XGBoost are relatively slow, taking around 1.7 seconds and 0.9 seconds, respectively.

#### **Insights:**

- KNN and Naive Bayes are the fastest, making them ideal for quick analysis.
- Logistic Regression, SGD, and Decision Tree offer moderate speed and complexity, suitable for balanced performance.
- SVM, while potentially more accurate, is the slowest and requires significant resources.
- Random Forest and XGBoost have relatively slower runtimes, but they may provide higher accuracy and robustness for complex datasets.
- Consider the trade-off between accuracy, complexity, and runtime when selecting a model based on the project's requirements.


####**Memory Utilized**

```python
import tracemalloc
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def get_model_memory_usage(model, X, y):
    tracemalloc.start()
    model.fit(X, y)
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / (1024 * 1024)  # Convert bytes to MB

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55022)
log_reg_memory = get_model_memory_usage(log_reg_model, ac07_X_train, ac07_y_train)
print(f"Logistic Regression memory usage: {log_reg_memory:.2f} MB")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55022)
svm_memory = get_model_memory_usage(svm_model, ac07_X_train, ac07_y_train)
print(f"SVM memory usage: {svm_memory:.2f} MB")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55022)
sgd_memory = get_model_memory_usage(sgd_model, ac07_X_train, ac07_y_train)
print(f"SGD memory usage: {sgd_memory:.2f} MB")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55022)
dt_memory = get_model_memory_usage(dt_model, ac07_X_train, ac07_y_train)
print(f"Decision Tree memory usage: {dt_memory:.2f} MB")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_memory = get_model_memory_usage(knn_model, ac07_X_train, ac07_y_train)
print(f"KNN memory usage: {knn_memory:.2f} MB")

# Naive Bayes
nb_model = GaussianNB()
nb_memory = get_model_memory_usage(nb_model, ac07_X_train, ac07_y_train)
print(f"Naive Bayes memory usage: {nb_memory:.2f} MB")

# Random Forest
rf_model = RandomForestClassifier(random_state=55022)
rf_memory = get_model_memory_usage(rf_model, ac07_X_train, ac07_y_train)
print(f"Random Forest memory usage: {rf_memory:.2f} MB")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55022)
xgb_memory = get_model_memory_usage(xgb_model, ac07_X_train, ac07_y_train)
print(f"XGBoost memory usage: {xgb_memory:.2f} MB")
```

### **Observations:**

- **Highest:** XGBoost has the highest memory usage, requiring 0.87 MB.
- **Lowest:** Naive Bayes has the lowest memory usage, requiring only 0.11 MB.
- **Moderate:** Logistic Regression, SVM, Decision Tree, and KNN have moderate memory usage, ranging from 0.22 MB to 0.34 MB.
- **Relatively Low:** SGD has relatively low memory usage, requiring 0.17 MB.
- **Relatively High:** Random Forest has relatively high memory usage, requiring 0.50 MB.

#### **Insights:**

- XGBoost has the highest memory usage, requiring careful resource management for large datasets.
- Random Forest also has relatively high memory usage compared to other models, which may limit its use in memory-constrained environments.
- Logistic Regression, SVM, Decision Tree, and KNN have moderate memory requirements, making them a balanced choice for most tasks.
- SGD and Naive Bayes have lower memory footprints, with Naive Bayes being the most efficient for memory-sensitive tasks.
- Consider both memory requirements and model performance when selecting the appropriate algorithm for your use case.
